Laboration 3 Linjär klassificering

Den här laborationen syftar till att undersöka hur linjär klassificering kan användas för att dela upp datapunkter i två klasser.
Jag implementerar en enkel linjär modell, klassificerar punkter och visualiserar resultaten för att förstå hur lutning och intercept påverkar separationen.

```python
import numpy as np
import matplotlib.pyplot as plt
from Functions import Functions 

# Läs in data
path = r"C:\Users\monza\Desktop\REPO\python-programming-M-NS-JILDESTRAND\Labs\Lab3\unlabelled_data.csv"
unlabelled = np.loadtxt(path, delimiter=",")
```

Jag använder funktionen Functions för att dela upp punkterna i de som ligger ovanför och under olika linjer.

```python
# Linje 1
x_above1, y_above1, x_below1, y_below1, above_idx1, below_idx1 = Functions(
    unlabelled[:,0], unlabelled[:,1], slope=-0.9, intercept=0.5
)

# Linje 2
x_above2, y_above2, x_below2, y_below2, above_idx2, below_idx2 = Functions(
    unlabelled[:,0], unlabelled[:,1], slope=-0.489, intercept=0
)

# Linje 3
x_above3, y_above3, x_below3, y_below3, above_idx3, below_idx3 = Functions(
    unlabelled[:,0], unlabelled[:,1], slope=-2, intercept=0.16
)

# Linje 4
x_above4, y_above4, x_below4, y_below4, above_idx4, below_idx4 = Functions(
    unlabelled[:,0], unlabelled[:,1], slope=800, intercept=-120
)

# Visa antal punkter ovanför/under varje linje
print("Linje 1:", len(x_above1), "ovanför,", len(x_below1), "under")
print("Linje 2:", len(x_above2), "ovanför,", len(x_below2), "under")
print("Linje 3:", len(x_above3), "ovanför,", len(x_below3), "under")
print("Linje 4:", len(x_above4), "ovanför,", len(x_below4), "under")
```

Här plottar jag datapunkterna tillsammans med linje 1
Röda punkter ligger ovanför linjen (klass 1) och blå punkter under linjen (klass 0).

```python
x_line = np.linspace(np.min(unlabelled[:,0]), np.max(unlabelled[:,0]), 100)

y_line1 = -0.9*x_line + 0.5
y_line2 = -0.489*x_line
y_line3 = -2*x_line + 0.16
y_line4 = 800*x_line - 120

plt.figure(figsize=(8,8), dpi=100)
plt.title("Klassificering av punkter")

# Rita linjer
plt.plot(x_line, y_line1, label='y = -0.9x + 0.5', color="black")
plt.plot(x_line, y_line2, label='y = -0.489x', color="green")
plt.plot(x_line, y_line3, label='y = -2x + 0.16', color="red")
plt.plot(x_line, y_line4, label='y = 800x - 120', color="blue")

# Rita punkter för linje 1
plt.scatter(x_above1, y_above1, c="red", label="Label 1 above")
plt.scatter(x_below1, y_below1, c="blue", label="Label 0 below")

plt.xlabel("X")
plt.ylabel("Y")
plt.ylim(-5, 5)
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

Analys

- Alla linjer delar upp punkterna nästan jämnt.

- Eftersom datapunkterna inte hade fördefinierade etiketter finns det många möjliga linjer som kan separera dem.

- Val av linje beror på kriterier såsom separation och balans.

- Huvudsyftet är att förstå principen för linjär klassificering och hur lutning (slope) och intercept påverkar klassindelningen.
